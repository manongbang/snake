```python
from datetime import datetime
import pandas_datareader.data as web
from f.portfolios import MarketOnClosePortfolio
from f.strategy.ma_cross import MovingAverageCrossStrategy
```

```python
for code in today_all_df['code']:
    symbol = code + ('.SS' if int(code[0]) >= 5 else '.SZ')
    try:
        bars = web.DataReader(symbol, 'yahoo', start=datetime(2016,8,29), end=datetime.now())
    except:
        # print 'data read error, ignore symbol:', symbol
        continue
    try:
        strategy_inst = MovingAverageCrossStrategy(symbol, bars)
    except:
        # print 'unsupport MAC, ignore symbol:', symbol
        continue
    signals = strategy_inst.generate_signals()
    last_positions = signals['positions'][-3:]
    trigger_buy = last_positions[last_positions == 1.0].count() > 0
    if trigger_buy:
        print stock_basics_df.loc[code]
```
