```python
import tushare as ts
```

```python
# refer to http://tushare.org/trading.html
# 用于获取所有股票代码
df = ts.get_today_all()
```

```python
import time
from datetime import datetime
import pandas as pd
from f.portfolios import MarketOnClosePortfolio
from f.strategy.ma_cross import MovingAverageCrossStrategy
```

```python
for symbol in df['code']:
    try:
        hist = ts.get_h_data(symbol, start=datetime(2016,8,29).strftime('%Y-%m-%d'), end=datetime.now().strftime('%Y-%m-%d'))
        time.sleep(1)
    except:
        print 'ignore symbol:', symbol
        continue
    bars = pd.DataFrame(index=hist.index)
    bars['Open'] = hist['open']
    bars['Close'] = hist['close']
    bars['Adj Close'] = hist['close']
    bars['High'] = hist['high']
    bars['Low'] = hist['low']
    bars['Volume'] = hist['volume']
    
    strategy_inst = MovingAverageCrossStrategy(symbol, bars)
    signals = strategy_inst.generate_signals()
    last_positions = signals['positions'][-3:]
    trigger_buy = last_positions[last_positions == 1.0].count() > 0
    if trigger_buy:
        print symbol
```
