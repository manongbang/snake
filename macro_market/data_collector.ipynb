```python
"""
    获取股票的基本面数据
    refer to http://tushare.org/fundamental.html
"""
import tushare as ts
import pickle

stock_basics = ts.get_stock_basics()  # 获取股票列表
```

```python
stock_codes = sorted([r[0] for r in stock_basics.iterrows()])
```

```python
def get_stock_data(years, seasons, func):
    data_set = {}
    for y in years:
        for s in seasons:
            year_season_str = '{y}-{s}'.format(y=y,s=s)
            try:
                data_set[year_season_str]=func(y,s)
                print 'get data:', year_season_str
            except:
                print 'fail to get data:', year_season_str
    return data_set
```

```python
fun_map = [
    ('reports', ts.get_report_data),
    ('profits', ts.get_profit_data),
    ('operations', ts.get_operation_data),
    ('growths', ts.get_growth_data), 
    ('debtpayings', ts.get_debtpaying_data), 
    ('cashflows', ts.get_cashflow_data),
]
```

```python
for name, func in fun_map:
    data = get_stock_data([1997+i for i in range(20)], [i+1 for i in range(4)], func)
    with open('tmp_data/stock_{name}_1997-2016.pkl'.format(name=name), 'w') as f:
        pickle.dump(data, f)
```

```python
def get_all_hist_data(code, year_interval=3):
    import pandas as pd
    from datetime import datetime, date
    
    def format_date(d):
        return d.strftime('%Y-%m-%d')
    
    proxies = ts.get_proxies(count=10)
    print proxies
    now_date = datetime.now().date()
    # 获取上市时间
    ipo_date = datetime.strptime(str(stock_basics.loc[code]['timeToMarket']), '%Y%m%d').date() if stock_basics.loc[code]['timeToMarket'] else date(2000,1,1)
    start_date = ipo_date
    end_date = date(start_date.year+year_interval, 1, 1)
    data_frames = []
    while now_date >= start_date:
        try:
            # print '**({s}, {e})**'.format(s=start_date, e=end_date)
            batch_df = ts.get_h_data(code, start=format_date(start_date), end=format_date(end_date), pause=3, proxies=proxies)
        except:
            continue
        data_frames.append(batch_df)
        start_date = end_date
        end_date = date(start_date.year+year_interval, 1, 1)
    return pd.concat(data_frames)
```

```python
df = get_all_hist_data('000333')
```

```python
import os
for code in stock_codes:
    file_name = 'tmp_data/stock_hist_{code}.pkl'.format(code=code)
    if os.path.exists(file_name):
        continue
    data = get_all_hist_data(code)
    with open(file_name, 'w') as f:
        pickle.dump(data, f)
```
